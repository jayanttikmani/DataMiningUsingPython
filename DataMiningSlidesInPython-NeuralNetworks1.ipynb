```python
import pandas as pd
#Loading Dataset in DataFrame
housePrices = pd.DataFrame.from_csv("C:\\Users\\Jayant\\Desktop\\DMPA\\PythonSlides\\HousePrices (1).csv", index_col=None)
housePrices
```

```python
#Defining Dependent and Independent varaibles for Multinomial Logistic Regression
import statsmodels.api as st
X = housePrices[["Price","SqFt","Brick"]]
X['Brick'] = X['Brick'].factorize()[0]
X = st.add_constant(X, prepend=False)
y = housePrices["Neighborhood"]
```

```python
#Fitting Multinomial Logistic Regression
model = st.MNLogit(y, X)
fit = model.fit()
print(fit.summary())
```

```python
#predicting probabilities for the above multinomial logistic regression
predictedProbab = fit.predict(X)
predictedProbab
```

```python
#predicting class for the above multinomial logistic regression
predictedClass = [0]*len(predictedProbab)
for i in predictedProbab.index.values:
    predictedClass[i] = np.argmax(np.max(predictedProbab.iloc[[i]]))
predictedClass
```

```python
#Train Test Split
np.random.seed(12345)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5)
```

```python
#Fitting the model on training set
model = st.MNLogit(y_train, X_train)
fit = model.fit()
print(fit.summary())
```

```python
#Predicting of test dataset
predictedProbab = fit.predict(X_test)
predictedProbab
```

```python
#predicting class of test dataset observations
predictedClass1 = [0]*len(predictedProbab)
for i in range(0,len(predictedProbab)):
    predictedClass1[i] = np.argmax(np.max(predictedProbab.iloc[[i]]))
predictedClass1
```

```python
#Confusion Matrix for test data
CM = pd.crosstab(np.asarray(y_test), np.asarray(predictedClass1),rownames=['Actual'],colnames=['Predicted'])
CM
```

```python
#calculating the accuracy for above confusion matrix
acc = (CM[0][0] + CM[1][1] + CM[2][2])/np.sum(np.sum(CM))
acc
```

```python
#Generating XOR data randomly
import numpy as np
np.random.seed(123457)
X1 = [np.random.randint(0,2) for i in range(147)]
X2 = [np.random.randint(0,2) for i in range(147)]
```

```python
#assigning dependent variable value based on the generated data
import pandas as pd
y = [0]*len(X1)
for i in range(len(X1)):
    if (X1[i] == X2[i]):
        y[i] = 0
    else:
        y[i] = 1
XORdata = pd.DataFrame(data={'X1': X1, "X2":X2, "y":y})
```

```python
#Running Logistic Regression on XOR data
import statsmodels.api as sm
import statsmodels.formula.api as smf
glmLog = smf.glm(formula = 'y ~ X1+X2', data=XORdata, family = sm.families.Binomial(sm.families.links.logit)).fit()
print(glmLog.summary())
```

```python
#Prediction on training data using fitted model
#Generating Predictions as probabilities
predictedProbability = glmLog.predict(XORdata[["X1","X2"]])
#Defining Cutoff
cutoff = 0.5
#Generating Prediction Class
import numpy as np
predicted = np.where(predictedProbability > 0.5, 1, 0)
predicted
```

```python
#Confusion Matrix for above logistic model
CM = pd.crosstab(np.asarray(XORdata.ix[:,"y"]),np.asarray(predicted), rownames=['Actual'], colnames=['Predicted'])
CM
```

```python
#Neural Network
from sklearn.neural_network import MLPClassifier
X = XORdata[["X1","X2"]]
y = XORdata["y"]
clf = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(2,))
clf.fit(X,y)
```

```python
#The ith element in the list represents the weight matrix corresponding to layer i.
clf.coefs_
```

```python
#The ith element in the list represents the bias vector corresponding to layer i + 1
clf.intercepts_
```

```python
#predicting XOR data using the fitted neural network model
predicted = clf.predict(X)
predicted
```

```python
#creating a confusion matrix for neural network model
CM = pd.crosstab(np.asarray(y),np.asarray(predicted), rownames=['Actual'], colnames=['Predicted'])
CM
```

```python
#Iris data neural networks
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets

iris = datasets.load_iris()
iris.data.shape, iris.target.shape
```

```python
#train test split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.4)
```

```python
X_train.shape, y_train.shape
```

```python
X_test.shape, y_test.shape
```

```python
#Initializing and fitting neural net on iris data
clf = MLPClassifier(solver='lbfgs', hidden_layer_sizes=(3,))
clf.fit(X_train,y_train)
```

```python
#The ith element in the list represents the weight matrix corresponding to layer i.
clf.coefs_
```

```python
#The ith element in the list represents the bias vector corresponding to layer i + 1
clf.intercepts_
```

```python
#making predictions on the test data using the fitted model
predicted = clf.predict_proba(X_test)
j = 0
predictedClass = [0]*len(predicted)
for i in predicted.tolist():
    predictedClass[j] = i.index((max(i)))
    j = j+1
    
predictedClass
```

```python
#creating a confusion matrix for neural network model
CM = pd.crosstab(np.asarray(y_test),np.asarray(predictedClass), rownames=['Actual'], colnames=['Predicted'])
CM
```
