##### Resampling.pptx

```python
import pandas as pd
#Loading Dataset in DataFrame
houseprices = pd.DataFrame.from_csv('C://Users//Jayant//Desktop//sem2//DMPA//PythonSlides//HousePrices (1).csv', index_col=None)
houseprices
```

#### Cross Validation on the Linear Regression Model

```python
#Simple Linear Regression

#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt

#Creating X and y variables for the linear regression
X = houseprices.iloc[:, 2:].values
y = houseprices.iloc[:, 1].values

#Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder = LabelEncoder()
X[:, 4] = labelencoder.fit_transform(X[:, 4])
X[:, 5] = labelencoder.fit_transform(X[:, 5])
onehotencoder = OneHotEncoder(categorical_features = [5], )
X = onehotencoder.fit_transform(X).toarray()

#Fitting simple linear regression to the training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X, y)
```

##### k-fold cross validation

```python
# Applying k (10) -fold cross validation | scoring='neg_mean_squared_error'
from sklearn.model_selection import cross_val_score
mse = cross_val_score(regressor, X, y, scoring='neg_mean_squared_error', cv=10)
```

```python
#mean squared error for 10 (k) fold cross validation
mse
```

```python
#The mean mse and the 95% confidence interval of the mse estimate:
print("Mean Squared Error: %0.2f (+/- %0.2f)" % (mse.mean(), mse.std() * 2))
```

#### 'scoring' techniques other than 'neg_mean_squared_error' that can be used on regressors while cross validation: ‘neg_mean_absolute_error’, ‘neg_median_absolute_error’, ‘r2’. An example of r2 scoring:

```python
# Applying k (10) -fold cross validation | scoring='r2'
from sklearn.model_selection import cross_val_score
r2 = cross_val_score(regressor, X, y, scoring='r2', cv=10)
```

```python
#r2 for 10 (k) fold cross validation
r2
```

```python
#The mean r2 and the 95% confidence interval of the r2 estimate:
print("R2: %0.2f (+/- %0.2f)" % (r2.mean(), r2.std() * 2))
```

##### Leave One Out Cross Validation (nothing but K-Fold with k = number of samples)

```python
# Applying Leave One Out (k = n) cross validation | scoring='neg_mean_squared_error'
from sklearn.model_selection import cross_val_score
mse = cross_val_score(regressor, X, y, scoring='neg_mean_squared_error', cv=len(X))
```

```python
#mean squared error for Leave One Out cross validation
mse
```

```python
#The mean mse and the 95% confidence interval of the mse estimate:
print("Mean Squared Error: %0.2f (+/- %0.2f)" % (mse.mean(), mse.std() * 2))
```

#### Cross Validation on the Logistic Regression model

```python
#Loading Dataset in DataFrame
beerPref = pd.DataFrame.from_csv('C://Users//Jayant//Desktop//sem2//DMPA//PythonSlides//Beer+Preferences.csv', index_col=None)
beerPref
```

```python
#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt

#Creating X and y variables for the linear regression
X = beerPref.iloc[:, [0,1,2,4]].values
y = beerPref.iloc[:, 5].values

#Fitting logistic regression to the training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X, y)
```

##### K-Fold Cross Validation

```python
# Applying k (10) -fold cross validation | scoring='accuracy'
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(classifier, X, y, scoring='accuracy', cv=10)
```

```python
#accuracies for 10 (k) fold cross validation
accuracies
```

```python
#The mean accracy and the 95% confidence interval of the accuracy estimate:
print("Accuracy: %0.2f (+/- %0.2f)" % (accuracies.mean(), accuracies.std() * 2))
```

```python
#Mean Error
#Mean Error
print("Error: %0.2f" % (1-accuracies.mean()))
```

#### 'scoring' techniques other than 'accuracy' that can be used on classifiers while cross validation: ‘average_precision’, ‘f1’, ‘neg_log_loss’, ‘precision’, ‘recall’, ‘roc_auc’.

##### Leave One Out Cross Validation (nothing but K-Fold with k = number of samples)

```python
# Applying LOO cross validation | scoring='accuracy'
from sklearn import cross_validation
loo = cross_validation.LeaveOneOut(len(y))

from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(classifier, X, y, scoring='accuracy', cv=loo)
```

```python
#accuracies for Leave One Out cross validation
accuracies
```

```python
#The mean accracy
print("Accuracy: %0.2f" % (accuracies.mean()))
```

```python
#Mean Error
print("Error: %0.2f" % (1-accuracies.mean()))
```
